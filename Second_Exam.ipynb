## 2.2 Add more info

```python
# Extract the year
ex['Year'] = ex['DATE'].ex.year

# Extract the Quarter
ex['Quarter'] = ex['DATE'].ex.quarter

# Extract the Month
ex['Month'] = ex['DATE'].ex.month

# Extract the Week
ex['Week'] = ex['DATE'].ex.week

# Extract the Weekday
ex['Weekday'] = ex['DATE'].ex.weekday

# Extract the Day
ex['Day'] = ex['DATE'].ex.day

ex.head()
```
